Handling Outliers

```python
def detect_outliers(data):
    outlier_list = []
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outlier_list.append(i)
    return outlier_list 

for i in ["age", "trestbps", "chol", "thalach", "oldpeak"]:
    outliers = detect_outliers(dataframe[i])
    print("Outliers in",i,"attribute :", outliers)
```

No outliers in age attribute

Outliers in trestbps attribute : 172, 174, 178, 178, 180, 180, 180, 192, 200

Outliers in chol attribute : 394, 407, 409, 417, 564

Outliers in thalach attribute : 71

Outliers in oldpeak attribute : 4.2, 4.2, 4.4, 5.6, 6.2

```python
def handle_outliers(data):
    tenth_percent = np.percentile(data, 10)
    ninentieth_percent = np.percentile(data, 90)
    
    print('10th per',tenth_percent)
    print('90th per',ninentieth_percent)

    return_tenth_percent= np.where(data<tenth_percent, tenth_percent, data)

    return_ninth_percent = np.where(return_tenth_percent>ninentieth_percent, ninentieth_percent, return_tenth_percent)
    return return_ninth_percent
    
    
for i in ['trestbps','chol','thalach','oldpeak']:
    dataframe[i]=handle_outliers(dataframe[i])

for i in ['trestbps','chol','thalach','oldpeak']:
    print(handle_outliers(dataframe[i]))
```

```python
from sklearn.model_selection import train_test_split 

feature_column = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']
X = dataframe[feature_column]
Y = dataframe.target
```

# Model Building

```python
## Logistic Regression
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, Y_train)
```

```python
Y_pred = log_reg.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(Y_test, Y_pred)
```

```python
lr_accuracy=round(metrics.accuracy_score(Y_test, Y_pred)*100,2)
print("\nAccuracy:",lr_accuracy,'%')
```

```python
metrics.confusion_matrix(Y_test, Y_pred )
```

```python
print(metrics.classification_report(Y_test, Y_pred))
```
